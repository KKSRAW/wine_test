# Case study on Wine test
- The dataset contains 1,143 rows and 13 columns, representing different chemical properties of wine samples along with their quality ratings.
- Fixed acidity:- Affects the wine's taste, structure, and aging potential.
- Volatile acidity :-High values can lead to an unpleasant smell and taste.
- Citric acid :-A wine with very low citric acid might taste dull, flat, or unbalanced because it lacks the sharp acidity that makes wine feel fresh.
- Residual sugar :-  Higher values mean sweeter wine
- Chlorides :- High levels can make the wine taste salty.
- Free sulfur dioxide:- to keep it fresh and prevent it from spoiling.  Too much SO₂ → Can give the wine a strong, unpleasant smell (like rotten eggs).
                        Too little SO₂ → The wine may spoil quickly and lose its flavor.
- Total sulfur dioxide :- Excess can negatively impact wine quality.
- Density:- Helps distinguish between dry and sweet wines.mass per unit volume of wine, affected by sugar and alcohol content.
- pH :- Affects wine stability and taste.
- Sulphates :- A type of salt that affects SO₂ levels and wine fermentation.
- Alcohol :- Higher alcohol generally leads to better quality perception
- Quality:- The target variable, indicating wine quality (score: 0-10).
- Id :-A unique identifier for each sample 



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv(r"E:\training\pandas\WineQT.csv")
df
```

```python

num_features = len(df.columns)
rows = (num_features // 3) + (num_features % 3 > 0)  # Dynamic row calculation

fig, axes = plt.subplots(nrows=rows, ncols=3, figsize=(15, rows * 3))
axes = axes.flatten()

colors = sns.color_palette("bright", num_features) 
for i, (col,color) in enumerate(zip(df.columns,colors)):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color=color)
    axes[i].set_title(col)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

corr_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Re-plot box plots with better spacing
fig, axes = plt.subplots(nrows=rows, ncols=3, figsize=(15, rows * 3))
axes = axes.flatten()

for i, col in enumerate(df.columns):
    sns.boxplot(y=df[col], ax=axes[i], color='orange')
    axes[i].set_title(col)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.countplot(x=df["quality"], palette="bright")
plt.title("Wine Quality Distribution", fontsize=14)
plt.xlabel("Quality Score")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Box Plot: Alcohol Content vs. Quality
plt.figure(figsize=(10, 6))
sns.boxplot(x="quality", y="alcohol", data=df, palette="bright")
plt.title("Alcohol Content vs. Quality", fontsize=14)
plt.xlabel("Quality Score")
plt.ylabel("Alcohol Content")
plt.show()
```

```python
# 4. Scatter Plot: Fixed Acidity vs. Volatile Acidity
plt.figure(figsize=(8, 6))
sns.scatterplot(x="fixed acidity", y="volatile acidity", hue="quality", data=df, palette="bright", alpha=0.7)
plt.title("Fixed Acidity vs. Volatile Acidity", fontsize=14)
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.show()
```
